```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("norbertelter/pcb-defect-dataset")

print("Path to dataset files:", path)
```

```python
import cv2
import matplotlib.pyplot as plt

sample_img_path = path+"/pcb-defect-dataset/train/images/l_light_01_missing_hole_01_1_600.jpg"
sample_img = cv2.cvtColor(cv2.imread(sample_img_path), cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 6))
plt.imshow(sample_img)
plt.axis('off')
plt.title("PCB DEFECT SAMPLE")
plt.show()
```

```python
import math

training_path = path+"/pcb-defect-dataset/train/images/"
pcb_types = ["missing hole", "open circuit", "short circuit", "spur", "spurious copper", "mouse bite"]
pcb_types_path = [training_path+"l_light_01_missing_hole_01_1_600.jpg", 
                  training_path+"l_light_01_open_circuit_01_1_600.jpg", 
                  training_path+"l_light_01_short_06_1_600.jpg", 
                  training_path+"l_light_01_spur_01_2_600.jpg", 
                  training_path+"l_light_01_spurious_copper_07_2_600.jpg",
                  training_path+"l_light_01_mouse_bite_20_1_600.jpg"]
pcb_types_img = [cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB) for path in pcb_types_path]

fig, ax = plt.subplots(2, 3, figsize=(10, 7))
for i in range(len(pcb_types_img)):
    ax[math.floor(i/3), i%3].imshow(pcb_types_img[i])
    ax[math.floor(i/3), i%3].axis('off')
    ax[math.floor(i/3), i%3].set_title(pcb_types[i])
    pass
```
